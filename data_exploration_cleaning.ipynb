```python
# Import dependencies
import pandas as pd
```

```python
# Load structures data and create DF
structures_df = pd.read_csv("../Resources/structures.csv")
print(structures_df.shape)
structures_df.head()
```

```python
# Review data types
dict(structures_df.dtypes)
```

```python
# Check for null values
dict(structures_df.isnull().sum())
```

```python
# Check for duplicates
structures_df.duplicated().sum()
```

```python
# Drop duplicates
structures_df = structures_df.drop_duplicates()
print(structures_df.shape)
structures_df.head()
```

```python
# Save clean structure data as CSV
structures_df.to_csv("../Resources/structures_clean.csv", index = False)
```

```python
# Load outcome data and create DF
outcomes_df = pd.read_csv("../Resources/sample-outcomes_LPedit.csv")
print(outcomes_df.shape)
outcomes_df.head()
```

```python
# Review data types
dict(outcomes_df.dtypes)
```

```python
# Rename "Unnamed: 20" column
outcomes_df = outcomes_df.rename(columns = {"Unnamed: 20": "EXTRA_COMMENTS"})
outcomes_df.head()
```

```python
# Convert percentage columns to numerical
## SOME OF THE "PERCENT_YIELD" COLUMNS HAVE #VALUE! IN THEM, WHICH NEEDS TO BE RESOLVED BEFORE THIS CONVERSION CAN BE COMPLETED
# outcomes_df["PERCENT_YIELD"] = outcomes_df["PERCENT_YIELD"].str.replace("%", "")
# outcomes_df["PERCENT_YIELD"] = outcomes_df["PERCENT_YIELD"].astype("float") / 100

outcomes_df["%_PURITY_(BY_LCMS)"] = outcomes_df["%_PURITY_(BY_LCMS)"].str.replace("%", "")
outcomes_df["%_PURITY_(BY_LCMS)"] = outcomes_df["%_PURITY_(BY_LCMS)"].astype("float") / 100

outcomes_df.head()
```

```python
# Check for null values
dict(outcomes_df.isnull().sum())
```

```python
outcomes_df = outcomes_df[outcomes_df["STRUCTURE_ID"].notnull()]
print(outcomes_df.shape)
outcomes_df.head()
```

```python
# Check for remaining null values
dict(outcomes_df.isnull().sum())
```

```python
# Save clean outcome data as CSV
outcomes_df.to_csv("../Resources/outcomes_clean.csv", index = False)
```
